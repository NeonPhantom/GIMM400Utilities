```
def WeightSum(numbers1, numbers2):
  
  #Sets the sum at 0
  sum = 0;

  #For each number in the numbers1 array, multiply each number in numbers1 by
  #its counterpart in numbers2.
  #For example, the 1st number in numbers1 is multiplied by the 1st number in
  #numbers2, and the 2nd numbers from each array are multiplied, and so on.
  for i in range(len(numbers1)):

    #Each set of multiplied numbers is added to the sum.
    sum = sum + (numbers1[i] * numbers2[i])
  
  #The weighted sum is printed.
  print(sum)


#This is used to enter integers into the 1st array of numbers.
num1 = [int(x) for x in input("enter in ints separated by white space: ").split()]

#This is used to enter the decimal counterparts (ex: 0.5) to the integers in
#the previous array. There needs to be the same amount of decimals as there are
#integers.
num2 = [float(x) for x in input("enter in floats separated by white space: ").split()]
WeightSum(num1, num2)
```

```
def StdDev(numbers):
  
  #Starts the 1st sum at 0.
  sum1 = 0

  #This calculates the sum of the numbers in the array.
  for i in range(len(numbers)):
    sum1 = sum1 + numbers[i]
    
  #This finds the mean of the aformentioned sum.
  mean = sum1 / len(numbers)
  
  #Starts the 2nd sum at 0.
  sum2 = 0

  for i in range(len(numbers)):
    #For each number in the array, subtract the mean from it and take square
    #that answer...
    randoNum = (numbers[i] - mean) ** 2
    #...And then add that to sum2.
    sum2 = sum2 + randoNum

  #Finally calculates the standard deviation using sum2.
  std = ((1 / len(numbers)) * sum2) ** (1 / 2)
  
  #Prints the standard deviation.
  print(std)

#This is used to enter integers into the array.
num1 = [int(x) for x in input("enter in ints separated by white space: ").split()]
StdDev(num1)
```
